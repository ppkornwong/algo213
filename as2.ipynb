```
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import pandas as pd
import datetime
```

```
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2021,3,1)

#stock
americanair = pdr.get_data_yahoo('AAL',start,end)
delta = pdr.get_data_yahoo('DAL',start,end)
unitedair = pdr.get_data_yahoo('UAL',start,end)

compareopen = pd.concat([americanair['Open'],delta['Open'],unitedair['Open']],axis = 1)
compareopen.columns = ['American air Open Price','Delta open price','United air open price']

scatter_matrix(compareopen)
```

```
startPTT = datetime.datetime(2011,3,1)
endPTT = datetime.datetime(2021,3,1)

PTTEP = pdr.get_data_yahoo('PTTEP.bk',startPTT,endPTT)

print(PTTEP['Volume'].argmax()) #print row of the max value of the column
print(PTTEP.iloc[PTTEP['Volume'].argmax()]) # print iloc for that row
```

```
startTesla = datetime.date(2014,1,1)
endTesla = datetime.date(2021,3,2)

Tesla = pdr.get_data_yahoo('tsla',startTesla,endTesla)
nday = endTesla - startTesla
count = countday(nday)
#print(nday)
#print(count)
closing = np.array([Tesla['Close']])
sumclose = np.sum(closing)
#print(sumclose)
movingave = sumclose/count
print(movingave)
```

```
starta = datetime.date(2014,1,1)
enda = datetime.date(2021,3,2)
apple = pdr.get_data_yahoo('aapl',starta,enda)
apple['return'] = ((apple['Close'] - apple['Open'])/ apple['Open'])*100

fig = plt.figure(figsize= (10,7))
ax = fig.add_axes([0, 0, 1, 1]) 
  
# Creating plot 
bp = ax.boxplot(apple['return']) 
  
# show plot 
plt.show() 
print(apple['return'])

print(apple['return'].median())

```

```
apple['return'].hist(bins=30,label = 'apple',alpha = 0.5) #bin=xaxis
plt.legend()
```
