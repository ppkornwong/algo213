```
#Sorrasak Promdum 62070504008 , Pidchakrit Tangwongsan 62070504011, Araya Kornwong 62070504015
import string
from collections import Counter
import time
start_time = time.time()
f = open('/content/501-2.txt', 'r')
content = f.read()
content1 = content.split() 
for i in range(len(content1)):
    if any([content1[i].startswith(c) for c in string.punctuation]):
        content1[i] = ' '
        if any([content1[i].startswith(c) for c in string.digits]):
            content1[i] = ' '
```

```
#create list
wlist = []

#use string to classify list that we want 
for i in range(len(content1)):
    if any([content1[i].startswith(c) for c in string.ascii_letters]):
        if any([content1[i].endswith(c) for c in string.punctuation]):
            wlist.append(content1[i].rstrip(string.punctuation).lower())
        else:
            wlist.append(content1[i].strip().lower())

x = chr(9) #ASCII of tab
#use counter
wordcounter = Counter(wlist)

sorted_by_count = wordcounter.most_common() #sorted by count

#create print of value and keys 
for i in range(500):#printing first 500

  key = sorted_by_count[i]
  key1 = str(key).split(',')[0].strip(string.punctuation)
  value = str(key).split(',')[1].strip(string.punctuation)
  print(key1 + x + value)

print("--- %s seconds ---" % (time.time() - start_time))
```
